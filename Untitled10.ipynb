```
F[x_,y_]:=18x^2-7x y+8y^2
G[x_,y_]:=x^2+y^2-2
L[x_,y_,l_]:=F[x,y]-l (G[x,y])
```

```
f1=D[L[x,y,l],x]
f2=D[L[x,y,l],y]
f3=D[L[x,y,l],l]
```

```
S=Solve[{f1==0,f2==0,f3==0},{x,y,l}]
```
