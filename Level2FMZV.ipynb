```
QMod[q_,p_]:=Module[{Aq,Vq},
Aq=Numerator[q];
Vq=Denominator[q];
If[GCD[Vq,p]!=1,0,Solve[Vq*x-Aq==0,x,Modulus->p][[1,1,2]]]
];

AMod[q_,p_]:=
If[IntegerQ[q]==True,Mod[q,p],QMod[q,p]];

Upper[n_]:=
If[IntegerQ[n]==True,n-1,Floor[n]];

Bernoulli[k_]:=If[k>=0,BernoulliB[k],0] ;

Z[k_]:=Table[AMod[Bernoulli[Prime[n]-k]/k,Prime[n]],{n,1,100}];

TruncatedMZV[list_,M_]:=Module[{l},
    l=Length[list];
    i=1;
     s=1;
     While[i<=l-1,
         s=Sum[(s*1)/(Subscript[m,i]^list[[i]]),{Subscript[m,i],1,Subscript[m,i+1]-1}];
         i++
        ];
     s=Sum[(s*1)/(Subscript[m,l]^list[[l]]),{Subscript[m,l],1,Upper[M]}]
    ];
    
FMZV[list_]:=
Table[AMod[TruncatedMZV[list,Prime[n]],Prime[n]],{n,1,100}];
L2FMZV[list_]:=
Table[AMod[TruncatedMZV[list,Prime[n]/2],Prime[n]],{n,1,100}];

Diff[list_]:=FMZV[list]-L2FMZV[list]
```
